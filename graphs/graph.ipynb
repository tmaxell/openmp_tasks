```python
#  
import matplotlib.pyplot as plt

#   
sizes = [100, 10000, 100000, 1000000]
time_with_reduction = [0.000301123, 0.000299215, 0.000653028, 0.00148916]
time_without_reduction = [0.000120878, 0.00180197, 0.056627, 0.713945]

#  
speedup = [time_without_reduction[i] / time_with_reduction[i] for i in range(len(sizes))]

#       
plt.figure(figsize=(10, 5))

#    
plt.subplot(1, 2, 1)
plt.plot(sizes, time_with_reduction, label='With Reduction', marker='o', color='b')
plt.plot(sizes, time_without_reduction, label='Without Reduction', marker='o', color='r')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Array Size (log scale)')
plt.ylabel('Execution Time (log scale) [seconds]')
plt.title('Execution Time: With vs Without Reduction')
plt.legend()

#  
plt.subplot(1, 2, 2)
plt.plot(sizes, speedup, label='Speedup', marker='o', color='g')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Array Size (log scale)')
plt.ylabel('Speedup')
plt.title('Speedup: Without / With Reduction')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#  2
import matplotlib.pyplot as plt
import numpy as np

#   
sizes = [10000, 100000, 1000000, 10000000, 100000000]  #  
times_with_reduction = [0.000513077, 0.00142717, 0.00687194, 0.070051, 1.66639]  #   
times_without_reduction = [0.00032115, 0.00111103, 0.00757194, 0.0706689, 2.29206]  #   

#  
speedup = np.array(times_without_reduction) / np.array(times_with_reduction)

#    
plt.figure(figsize=(12, 6))

#    
plt.subplot(1, 2, 1)
plt.plot(sizes, times_with_reduction, label='With Reduction', marker='o', linestyle='-', color='b')
plt.plot(sizes, times_without_reduction, label='Without Reduction', marker='o', linestyle='-', color='r')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Vector Size (log scale)')
plt.ylabel('Time (seconds) (log scale)')
plt.title('Execution Time vs. Vector Size')
plt.legend()

#  
plt.subplot(1, 2, 2)
plt.plot(sizes, speedup, marker='o', linestyle='-', color='g')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Vector Size (log scale)')
plt.ylabel('Speedup (log scale)')
plt.title('Speedup vs. Vector Size')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

#  (  N,       , )
N_values = [10000, 100000, 1000000, 100000000, 1000000000]
sequential_times = [0.000140905, 0.00272989, 0.0158801, 0.789132, 7.59109]
parallel_times = [0.000625134, 0.000916004, 0.00463104, 0.185301, 3.02436]
speedups = [0.2254, 2.98022, 3.42906, 4.25865, 2.50998]

#     
fig, ax = plt.subplots(1, 2, figsize=(14, 6))  # 1 , 2 

#   
ax[0].plot(N_values, sequential_times, label="Sequential Time", color='blue', marker='o')
ax[0].plot(N_values, parallel_times, label="Parallel Time", color='red', marker='s')

#    
ax[0].set_xlabel("Number of Subdivisions (N)")
ax[0].set_ylabel("Execution Time (seconds)")
ax[0].set_xscale('log')
ax[0].set_yscale('log')  #          N
ax[0].set_title("Execution Time Comparison (Sequential vs Parallel)")
ax[0].legend()

#  
ax[1].plot(N_values, speedups, label="Speedup", color='green', marker='^')

#    
ax[1].set_xlabel("Number of Subdivisions (N)")
ax[1].set_ylabel("Speedup (sequential / parallel)")
ax[1].set_xscale('log')
ax[1].set_title("Speedup (Sequential vs Parallel)")
ax[1].legend()

#  
plt.tight_layout()  #   
plt.show()
```

```python
# 4 
import matplotlib.pyplot as plt

#  
matrix_sizes = ["100x100", "1000x1000", "10000x10000"]
sequential_times = [0.0000870228, 0.00606894, 0.413307]
parallel_times = [0.00161386, 0.00155687, 0.075141]
speedup = [seq / par for seq, par in zip(sequential_times, parallel_times)]

#  
fig, ax1 = plt.subplots()

#   
ax1.set_xlabel("Matrix Size")
ax1.set_ylabel("Execution Time (s)", color="tab:blue")
ax1.plot(matrix_sizes, sequential_times, label="Sequential Time", marker="o", color="tab:blue", linestyle="-")
ax1.plot(matrix_sizes, parallel_times, label="Parallel Time", marker="o", color="tab:cyan", linestyle="--")
ax1.tick_params(axis="y", labelcolor="tab:blue")
ax1.legend(loc="upper left")

#  
ax2 = ax1.twinx()
ax2.set_ylabel("Speedup", color="tab:red")
ax2.plot(matrix_sizes, speedup, label="Speedup", marker="s", color="tab:red", linestyle="-.")
ax2.tick_params(axis="y", labelcolor="tab:red")
ax2.legend(loc="upper right")

#     
plt.title("Performance Analysis: Sequential vs Parallel (OpenMP)")
fig.tight_layout()
plt.show()
```
